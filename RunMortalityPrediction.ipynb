```python
from sklearn.utils.validation import Parameter
import itertools
import subprocess
def grid_search_model_tuning(parameters):
    # Define the parameter grid
    param_grid = {
        'withCCS': [0, 1],  # Add or remove CCS feature
        'hiddenDimSize': [50, 100],  # Hidden layer sizes to try
        'batchSize': [25, 50, 75],  # Batch sizes to try
        'nEpochs': [50, 100, 150],  # Number of epochs to try
        'lr': [0.001, 0.01, 0.1],  # Learning rates to try
        'dropOut': [0.25, 0.5]  # Dropout probabilities to try
    }

    # Generate all combinations of parameter values
    param_values = [param_grid[param] for param in parameters]
    combinations = list(itertools.product(*param_values))

    # Perform grid search
    best_score = None
    best_params = None

    for combination in combinations:
        # Create a dictionary of parameter settings
        params = dict(zip(parameters, combination))

        run_model(params)

def run_model(params):
    command = [
        'python',
        '02_FFN_mortality.py',
        f'--withCCS={params["withCCS"]}',
        f'--hiddenDimSize={params["hiddenDimSize"]}',
        f'--batchSize={params["batchSize"]}',
        f'--nEpochs={params["nEpochs"]}',
        f'--lr={params["lr"]}',
        f'--dropOut={params["dropOut"]}'
    ]

    result = subprocess.run(command, capture_output=True, text=True)
    output_lines = result.stdout.strip().split('\n')
    last_line = output_lines[-1].strip()

    with open('params-out.txt', 'a') as f:
        f.write(f'{last_line}\n')
        f.write(f'Params: {params}\n')
        f.write('---\n')

    return last_line


parameters = ['withCCS', 'hiddenDimSize', 'batchSize', 'nEpochs', 'lr', 'dropOut']
grid_search_model_tuning(parameters)
```
